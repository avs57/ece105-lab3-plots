```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.

import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(seed=5140)  # replace 1234 with your last 4 digits

sensor_a = rng.normal(loc=25.0, scale=3.0, size=200)
sensor_b = rng.normal(loc=27.0, scale=4.5, size=200)
timestamps = rng.uniform(low=0.0, high=10.0, size=200)

print(sensor_a.shape)
print(sensor_b.shape)
print(timestamps.shape)
```

```python
# Scatter plot: sensor reading (y-axis) vs timestamp in seconds (x-axis).
# Color Sensor A points blue and Sensor B points orange.
# Add axis labels with units, a title, and a legend.
```

```python
plt.figure()

plt.scatter(timestamps, sensor_a, color='blue', label='Sensor A')
plt.scatter(timestamps, sensor_b, color='orange', label='Sensor B')

plt.xlabel("Time (s)")
plt.ylabel("Temperature (°C)")
plt.title("Sensor Readings vs Time")

plt.legend()

plt.show()
```

```python
# Overlaid histogram of Sensor A and Sensor B temperature distributions.
# Use 30 bins, alpha=0.5 for transparency so both distributions are visible.
# Add vertical dashed lines at each sensor's mean.
# Include a legend labeling each sensor.
```

```python
plt.figure()

plt.hist(sensor_a, bins=30, alpha=0.5, label='Sensor A')
plt.hist(sensor_b, bins=30, alpha=0.5, label='Sensor B')

mean_a = np.mean(sensor_a)
mean_b = np.mean(sensor_b)

plt.axvline(mean_a, color='blue', linestyle='dashed', linewidth=2)
plt.axvline(mean_b, color='orange', linestyle='dashed', linewidth=2)

plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.title("Temperature Distribution of Sensors")

plt.legend()

plt.show()
```

```python
# Side-by-side box plot comparing Sensor A and Sensor B distributions.
# Label x-axis with sensor names, y-axis with "Temperature (deg C)".
# Add a horizontal dashed line at the overall mean of both sensors combined.
```

```python
plt.figure()

plt.boxplot([sensor_a, sensor_b], labels=['Sensor A', 'Sensor B'])

overall_mean = np.mean(np.concatenate([sensor_a, sensor_b]))

plt.axhline(overall_mean, color='red', linestyle='dashed', linewidth=2, label='Overall Mean')

plt.ylabel("Temperature (°C)")
plt.title("Box Plot of Sensor Distributions")

plt.legend()

plt.show()
```

```python
plt.figure()

plt.boxplot([sensor_a, sensor_b], tick_labels=['Sensor A', 'Sensor B'])

overall_mean = np.mean(np.concatenate([sensor_a, sensor_b]))

plt.axhline(overall_mean, color='red', linestyle='dashed', linewidth=2, label='Overall Mean')

plt.ylabel("Temperature (°C)")
plt.title("Box Plot of Sensor Distributions")

plt.legend()

plt.show()
```
